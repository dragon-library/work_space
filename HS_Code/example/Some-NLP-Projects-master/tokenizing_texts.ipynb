```
import pandas as pd
```

```
products = pd.read_csv('amazon_baby.csv')
```

```
products.head()
```

```
rate = {1: 'Very Poor', 2: 'Poor', 3: 'Neutral', 4: 'Good', 5: 'Very Good'}
products['Opinion'] = products['rating'].map(rate)
```

```
products.head()
```

```
products['name_review'] = products['name'] + products['review']
```

```
products.head()
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
tokenizer = Tokenizer(oov_token='<OOV>')
```

```
sentences = []
for row in products['name_review']:
  row = str(row)
  sentences.append(row)
```

```
tokenizer.fit_on_texts(sentences)
```

```
word_index = tokenizer.word_index
print(len(word_index))
```

```
sequences = tokenizer.texts_to_sequences(sentences)
padded = pad_sequences(sequences, padding='post')
print(padded[0])
print(padded.shape)
```

```
label_tokenizer = Tokenizer()
label_tokenizer.fit_on_texts(products['Opinion'])
```

```
label_word_index = label_tokenizer.word_index
label_seq = label_tokenizer.texts_to_sequences(products['Opinion'])
```

```
print(label_seq)
print(label_word_index)
```
