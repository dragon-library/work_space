```python
import glob
import codecs
import numpy
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, f1_score


pd.set_option('display.max_rows', 20, 
              'display.max_columns', 100, 
              'display.float_format', '{:,.2f}'.format)
```

### Getdata from database

```python
def range_txt(txt):
    
    if len(txt) == 12:
        tariff = txt[4:10]
        return tariff
    elif len(txt) == 11:
        tariff = '0' + txt[4:9]
        return tariff
    else:
        tariff = txt[0:6]
        return tariff
```

```python
def manage_tariff(data):
#   data['length'] = data['heading'].apply(len)
  #  data['heading'] = data.heading.apply(range_txt)
    
    db = []
    for i in range(len(data)):
    
     
    
        db.append({'descript': data.product_desc[i] ,'class':data.heading[i] })
    
    return db
```

```python
 import pyodbc
    
def get_database(commands):
    conn = pyodbc.connect("Driver={SQL Server Native Client 11.0};"
                      "Server=YO-PC\SQLEXPRESS2016;"
                      "Database=ECS_DATA2;"
                      "Trusted_Connection=yes;")

    cursor = conn.cursor()
    database = pd.read_sql_query(commands, conn)
    database = database[['TariffCode', 'PdtDescriptionE']]
    database.rename(columns={'TariffCode' : 'tariff_code'}, inplace = True)
    database.rename(columns={'PdtDescriptionE' : 'product_desc'}, inplace = True)
    
    return database
```

```python
commands = '''  

SELECT * FROM Deci_InvoiceD
WHERE  left(refno,1)='D'

'''

deci = get_database(commands)
deci
```

```python
reads =pd.read_excel('data/deci.xlsx',sheet_name='Sheet1', parse_string =['heading'])
reads['heading'] = reads['heading'].map('{:04}'.format)
deci = reads[['heading','product_desc']]
deci
```

```python
reads =pd.read_excel('data/decx.xlsx',sheet_name='Sheet1', parse_string =['heading'])
reads['heading'] = reads['heading'].map('{:04}'.format)
decx = reads[['heading','product_desc']]
decx
```

```python
print(len(books))
concats = pd.concat([books,deci, decx], ignore_index=True)
concats
```

```python
concats['product_desc'] = concats['product_desc'].str.lower()
concats
```

```python
data = manage_tariff(concats)
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df['descript'].values
```

```python
#แต่จริงแล้วดึงมาเฉพาะ field text

arr_text=df.iloc[:]['descript'].values
arr_text
```

```python
#CountVectorizer ทำ tokenize

vect = CountVectorizer(stop_words='english', lowercase=True)
vect
```

```python
#เริ่มตัดคำ

count_train = vect.fit(arr_text)
count_train
```

```python
#เริ่มตัดคำ

count_train = vect.fit(arr_text)

#ดู vocabulary_

vect.vocabulary_
```

```python
#ทำ transform ให้อยู่ในรูปแบบ One Hot Encoding

transformer = vect.transform(arr_text)
transformer
```

```python
#ดูผลลัพท์ matrix ที่ได้โดย array จะประกอบด้วย context ใน 1 paragraph หากมี word ใน context ให้ค่าเป็น 1 และหากไม่มี word ใน context ให้ค่าเป็น 0

print(transformer.toarray())
```

```python
# เตรียม tf-idf

from sklearn.feature_extraction.text import TfidfTransformer
# Config tf-idf

tfidf_transformer = TfidfTransformer(smooth_idf=False,).fit(transformer)

#show ค่า idf

tfidf_transformer.idf_
```

```python
from sklearn.naive_bayes import MultinomialNB

#เตรียม Class ในรูปแบบ array เหมือน arr_text

arr_class=df.iloc[:]['class'].values
arr_class
```

```python
#เตรียม train จาก transformer

messages_tfidf = tfidf_transformer.transform(transformer)
print (messages_tfidf.shape)
```

```python
messages_tfidf.toarray()
```

```python
#สร้าง Model

detect_model = MultinomialNB().fit(messages_tfidf,arr_class)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split( messages_tfidf,arr_class, test_size=0.4 , random_state=0)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

detect_model =MultinomialNB().fit(x_train,y_train)
#detect_model = MultinomialNB().fit(messages_tfidf,arr_class)
```

```python
print(detect_model.score(x_test, y_test))
```

```python
#การ Predict
#สร้าง ตัว test
#จะ test ด้วยคำว่า london bridge is falling down

arr_text[200]
```

```python
t0=vect.transform([arr_text[200]])

#แปลงเป็น vector

t0.toarray()
```

```python
#นำไป Predict

print ('Predicted: ',detect_model.predict(messages_tfidf[200]) )
```

```python
#shot copy

def model_predict(txt):
    vectmp = vect.transform([txt])
    tftmp = tfidf_transformer.transform(vectmp)
    return detect_model.predict(tftmp)
```

```python
txt = 'potatoes'
d = model_predict(txt)
print ('Predicted: ',d )
```

```python
#เห็นว่าคำทำนายออกมาเป็น london ตรงกับ Expected
#นำ data ทั้งหมดที่ train ไป test มันก็ควรจะตรง(Expected คือ 100%)

all_predictions = detect_model.predict(messages_tfidf)
print(all_predictions)
```

```python
#ดูค่า Performance

from sklearn.metrics import classification_report
print (classification_report(arr_class, all_predictions))
```
